```python
# outlier test
import matplotlib.pyplot as plt
import seaborn as sns

# Adjust the figure size
plt.figure(figsize=(10, 8))

# Creating box plots for each numerical column
sns.boxplot(data=X_scaled_num)
plt.xticks(rotation=90)  # Rotate x-labels for better visibility
plt.show()

def find_outliers_indices(df):
    outliers_dict = {}
    # For each feature find the data points with extreme high or low values
    for feature in df.columns:
        Q1 = np.percentile(df[feature], 25)
        Q3 = np.percentile(df[feature], 75)
        IQR = Q3 - Q1
        
        # Determine a list of indices of outliers for feature
        outlier_indices = df[(df[feature] < (Q1 - 1.5 * IQR)) | 
                             (df[feature] > (Q3 + 1.5 * IQR))].index
        
        # Append the found outlier indices for feature to the dictionary
        outliers_dict[feature] = {
            'indices': outlier_indices.tolist(),
            'num_outliers': len(outlier_indices)
        }
        
    return outliers_dict

# Getting outliers indices for each column
outliers_indices = find_outliers_indices(X_scaled_num)

# Printing the number of outliers and their indices for each column
for column, details in outliers_indices.items():
    print(f"Column '{column}' has {details['num_outliers']} outliers: {details['indices']}")
```

```python
#Feature Selection: Random Forest
from sklearn.ensemble import RandomForestClassifier 

# Fit the model
randomforest = RandomForestClassifier(random_state=0)
model_rf = randomforest.fit(X_train, y_train)

# Get feature importances and sort them
feature_importances = pd.Series(model_rf.feature_importances_, index=X.columns).sort_values(ascending=False)

# Select top 30 features
top_30_features = feature_importances.head(30)

# Plot the top 30 features
top_30_features.plot(kind='bar', figsize=(14,6))
plt.title("Top 30 Predictors based on Feature Importance")
plt.ylabel("Importance")
plt.show()
```

```python

## Principal component analysis
from sklearn.decomposition import PCA
pca = PCA(n_components=50)
pca.fit(X_scaled) #requires standardization

pca.explained_variance_ratio_

pca.components_

# Using elbow method to select number of components
import matplotlib.pyplot as plt
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```
